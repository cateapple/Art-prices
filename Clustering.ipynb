# k-means

## PCA

```python
from sklearn.decomposition import PCA

# Create a PCA model to reduce our data to 2 dimensions for visualisation
pca = PCA(n_components=2)
pca.fit(X)

# Transfor the scaled data to the new PCA space
X_reduced = pca.transform(X)
```

```python
pca.explained_variance_ratio_
```

We keep around 77% of the variance explained here with 2 dimensions.

## k-means

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8).fit(X_reduced)
```

```python
#Plot the clusters obtained using k means
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
```

## Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=8).fit(X_reduced)
#Plot the clusters obtained using Agglomerative Clustering
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=agg.labels_, s=50, cmap='viridis')
```
